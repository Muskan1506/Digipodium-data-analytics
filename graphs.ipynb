```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall =[17, 10, 10, 10, 26, 59]
months =['jan','feb','mar','apr','may','jun']
```

```python
plt.style.available
```

```python
plt.bar(months, rainfall, color='#1f77b4') 
```

```python
plt.style.use('ggplot')  #theme
```

```python
plt.pie (
rainfall,
labels=months,
autopct='1%1.1f%%'
)
plt.show()
```

```python
#donut 
plt.pie(

    rainfall,
    labels = months,
    autopct='%.1f%%',
    wedgeprops= {'width':.4, 'edgecolor':'#6666ff','linewidth':2},
    pctdistance=.8
)
plt.show()

```

```python
plt.boxplot(rainfall,
           vert=False,
            showmeans=True,
             meanline=True,
             meanprops={'color':'red'}
             )
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(5, 5))  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], label='kuch bhi') # Plot some data on the axes.
ax.plot([1, 2, 3, 4],[2, 3, 4, 1] ,label='kch bhi')
ax.grid()
ax.legend()
```

```python
t= np.arange(-10, 10, .2)
y=np.sin(t)
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(t, y, label='sin wave')
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```

```python
t= np.arange(-10, 10, .2)
y=np.sin(t)
y2=np.cos(t)
fig,ax=plt.subplots(figsize=(10,3))
ax.plot(t, y, label='sin wave',
        color='#88ff99',marker='o',markersize=7, markerfacecolor='green',
        alpha=.4)
ax.plot(t, y2, label='cosine wave',
        color='#ffff99',marker='o',markersize=8, markerfacecolor='red',
        alpha=.7)
ax.legend()
ax.set_xlabel('time')
ax.set_ylabel('amplitude')
```
